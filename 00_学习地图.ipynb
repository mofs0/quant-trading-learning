## 入门前的 60 秒自测

下面 10 道题，先不要查资料，看你能答对几道。这会告诉你应该从哪一节开始。

**计分规则：**

- 0-3 分：从 Notebook 01 开始，按部就班学
- 4-6 分：可以快速过 01，重点看 02-03
- 7-9 分：直接看 03 求职部分 + 04 起的 Web3 内容
- 10 分：你不需要这个手册，去面试吧

> 答案在本节末尾。

**题目：**

1. 一支股票年化收益 20%、年化波动 15%，无风险利率 3%，Sharpe 是多少？
2. 多因子模型中，IC（Information Coefficient）的定义是什么？
3. 什么叫"未来函数"？举一个例子。
4. 对一组时间序列做交叉验证，普通 KFold 有什么问题？
5. Black-Scholes 公式里，Vega 是什么？
6. Python 里 `df.groupby('date')['close'].rolling(20).mean()` 这行代码常见的 bug 是什么？
7. 私募和公募的核心区别是什么？
8. 区块链的"不可篡改"靠什么实现？
9. Solidity 里 `view` 和 `pure` 函数有什么区别？
10. AMM（自动做市商）里 Uniswap V2 的恒定乘积公式是什么？

试着写出你的答案再往下翻 ↓

**参考答案：**

1. (0.20 - 0.03) / 0.15 ≈ **1.13**
2. 因子值与未来收益的**横截面相关系数**（通常用 Spearman）
3. 用了 T 时刻之后才能获得的信息做 T 时刻的决策。例：用当日收盘价决定当日开盘下单。
4. **泄露未来信息**——训练集可能包含测试集之后的样本。要用 walk-forward / time-series CV。
5. 波动率每变化 1%，期权价格的变化量。
6. 索引混乱——`groupby` 之后的 `rolling` 不会自动按 group 内的时间排序，要先 `sort_values('date')`，或用 `groupby().rolling()` 后 reset_index。
7. 公募面向公众，门槛低（1 元起），监管严，杠杆限制多；私募面向合格投资者（100 万起），策略灵活，业绩报酬制。
8. **密码学哈希链** + **分布式共识**。任何修改会改变哈希，多节点验证使篡改成本极高。
9. `view` 可以读链上状态但不能修改；`pure` 既不能读也不能写，只能用参数计算。
10. **x · y = k**（池子里两种代币数量的乘积恒定）
