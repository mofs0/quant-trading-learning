### 7.2 高频概率题（必会）

**Q1：两人轮流抛硬币，先出正面的赢，问先抛者赢的概率？**

答：设先抛者赢的概率为 P。第一次正面赢（概率 1/2）；第一次反面，相当于对方变成先抛者，自己赢的概率变成 1 - P。
所以 P = 1/2 + 1/2 × (1 - P)，解得 P = 2/3。

**Q2：袋子里 5 个红球 3 个蓝球，不放回取，问取出第 4 个是红球的概率？**

答：直接 5/8。"位置交换"思想：任何位置抽到红球的概率都是 5/8。

**Q3：均匀分布 U(0,1) 取 100 个数，最大值的期望？**

答：n/(n+1) = 100/101。证明：$\max$ 的 CDF 是 $x^n$，PDF 是 $nx^{n-1}$，期望 = $\int_0^1 x \cdot nx^{n-1} dx = n/(n+1)$。

**Q4：一个老虎机，每次拉概率 p 中奖。问期望拉多少次首次中奖？**

答：几何分布，期望 = 1/p。

**Q5：从一副 52 张扑克随机抽 5 张，问全是不同花色的概率？**

答：4 种花色选 4 种（C(4,4)=1），但还差一张要重复。所以是恰好 4 种花色或重复。
精确算："4 种花色都有"的概率 = (52 × 39 × 26 × 13) / (52 × 51 × 50 × 49) × C(5,4)/... 算起来麻烦，建议手撸一遍。

**框架（STAR + 反思）：**
- **Situation**：项目动机，为什么做
- **Task**：具体任务和目标
- **Action**：你的方法和关键决策
- **Result**：量化的结果
- **Reflection**：局限性、可以怎么改进

**示例答案：**
> 我在 2024 年做了一个 A 股多因子选股项目，目标是构建一个相对沪深 300 的指增策略。
>
> 数据用 Tushare Pro，覆盖 2018-2023 共 5 年，约 4000 只股票。我构造了 18 个因子，包括 9 个量价类（动量、反转、波动率、量价相关性）、6 个基本面类（ROE、PB、Net Margin 等）、3 个分析师预期类。
>
> 关键决策有三个：第一，所有因子都做了行业市值中性化，避免 alpha 来自风格漂移；第二，用 IC 加权而不是等权合成，并用 60 日滚动 IC 防止未来函数；第三，组合优化用 cvxpy，做了行业偏离约束。
>
> 结果是 walk-forward 测试期年化收益 19%，年化波动 18%，Sharpe 1.05，最大回撤 22%。相对沪深 300 超额收益年化约 11%。
>
> 不足有几点：一是没有考虑停牌、ST 的特殊处理，可能高估了流动性；二是因子合成方法相对粗糙，应该尝试 LSTM 或 XGBoost；三是没做严格的 multiple testing 修正，p-hacking 风险存在。下一步我打算引入 PIT 财务数据，并用 ML 方法合成。

这种答案有数字、有思考、主动提不足，**比说"做了一个还不错的策略"强 10 倍**。

### 8.2 知识深度类

**Q：你怎么防止策略过拟合？**

> 过拟合是量化研究的头号敌人。我会从五个角度防：
> 1. **样本外验证**：严格的 walk-forward，训练集只用 t 时刻之前的数据
> 2. **参数节制**：因子数 < 样本数 / 50，参数尽量少
> 3. **多重检验修正**：试了多少个想法心里有数，用 Bonferroni 或 White Reality Check
> 4. **经济学逻辑兜底**：不能从直觉解释的信号不要——比如"周三的动量比周二好"这种就是数据挖出来的伪信号
> 5. **稳健性测试**：在不同市场（A股 / 美股）、不同时间窗口、不同参数附近都要验证

**Q：你怎么评价一个策略好不好？**

> 不只是 Sharpe。我会看：
> 1. 风险调整后收益：Sharpe、Sortino、Calmar
> 2. 容量：单笔交易冲击成本测试，问"加大 10x 资金 Sharpe 会变多少"
> 3. 与已有策略相关性：相关 > 0.7 没价值，需要找出 alpha 来源
> 4. 回撤特征：深度、频率、恢复期分布
> 5. 因子归因：超额收益是因子贡献还是 alpha
> 6. 不同市场环境表现：牛市、熊市、震荡市分别 Sharpe 多少

**Q：因子失效你怎么办？**

> 首先要诊断是"暂时回撤"还是"结构性失效"。前者用滚动 IC、ICIR 衰减监控判断；后者表现为 IC 长期为零甚至反向。
>
> 短期回撤：保持仓位，不轻易调整，但加强监控
> 结构性失效：拆解原因。可能是：
> - 市场结构变化（如注册制后小盘股弹性变化）
> - 因子拥挤（被太多机构使用）
> - 数据问题（如 Wind 调整了某个字段定义）
>
> 应对方式包括：因子升级（加条件因子）、组合层面降权、新因子接力。

### 8.3 思维 / 智力题

**Q：你在两支基金中选，A 历史 5 年年化 15%、Sharpe 1.5，B 历史 3 年年化 25%、Sharpe 2.0，选哪个？**

> 不能光看数字。我会问：
> - 两个基金策略类型一样吗？同等可比吗？
> - B 的 3 年里有没有特殊市场环境（如 2020 大涨）？
> - 两支基金规模如何？容量是否接近上限？
> - 团队稳定性？历史回撤是否被刻意压缩？
> - 费率结构如何？业绩报酬怎么算？
>
> 一般倾向于 A（更长历史、更稳健），但要排查所有问题。这道题考的是"你会不会被表面数字骗"。

**Q：解释一下你最近读的一篇量化论文。**

提前准备 1-2 篇你读懂的论文。不需要顶刊，**比较简单但你能讲清楚原理**的论文是最好的选择。

例如：
- Jegadeesh & Titman (1993) Momentum
- Fama-French 三因子模型
- López de Prado 的 Triple Barrier Labeling
- Avellaneda-Stoikov 做市模型

### 8.4 你的问题（反向提问）

面试官问"你有什么问题"时，**一定要问**。问什么很有讲究：

✅ 好问题：
- "团队的研究流程是什么样的？一个新 idea 从提出到上线大概多久？"
- "新人入职会被怎么培养？前 6 个月主要会做什么？"
- "团队最近半年最让你兴奋的研究方向是什么？"
- "团队对工具链的投入怎么样？研究效率和工程基础设施哪个是瓶颈？"

❌ 差问题：
- "工资多少？"（HR 谈，别问技术面）
- "加班严重吗？"（行业内都加，问了显得不专业）
- "你们策略具体是什么？"（人家不会告诉你）

## 9. 实习是最重要的敲门砖

> **80% 的量化全职 offer 来自实习转正。**

### 9.1 时间线

```
研一上 (9-12月)   → 学基础、做项目
研一下 (1-6月)    → 投暑期实习 / 日常实习
研一暑假 (7-9月)  → 暑期实习（最重要）
研二上 (9-12月)   → 留用 + 秋招
研二下 (1-6月)    → 春招补录
研二暑假          → 实习转正 / 拿 offer
研三上            → 入职
```

### 9.2 找实习的渠道（按效率排序）

1. **内推**：通过学长学姐 / 老师 / LinkedIn 找人内推。**效率最高**
2. **公司官网**：很多私募只在自己官网招聘
3. **牛客网 / 一亩三分地**：行业讨论 + 内推渠道
4. **微信群**：量化求职群、各校量化协会
5. **BOSS 直聘**：偶尔有，但鱼龙混杂
6. **猎头**：到了高级岗位才用

### 9.3 实习选公司的优先级

**优先选**：
- 能真正做研究的小公司 > 只能跑数据的大公司
- 有具体导师带的 > 放养式的
- 留用率高的 > 留用率低的（这个问问学长学姐就知道）
- 给你的方向是你未来想做的方向

**避坑**：
- 那些一招就是几十个实习生、没人管的"廉价劳动力"型
- 让你做"用 ChatGPT 写策略"这种没含金量的活
- 强制签长期、不签实习证明的

### 9.4 实习中怎么表现

1. **第一个月**：把工具链熟悉好，问问题别怕"显得蠢"
2. **第二个月**：找一个能独立负责的小项目，独立做完
3. **第三个月**：开始能贡献新 idea，主动找导师讨论
4. **离开前**：写一份漂亮的总结报告，把你做过的所有工作 list 出来

## 10. 不同背景的人，怎么规划

### 10.1 我是金融硕士，没编程基础

- **第一年**：Python 死磕（用 leetcode-python 刷 100 题）；多因子项目 1 个
- **第二年**：实习 + 第二个项目（CTA 或 ML）
- 路径建议：先去公募 / 券商资管，再考虑跳私募
- **避免**：直接投头部私募研究员，会被秒拒；先去中型私募刷经验

### 10.2 我是计算机硕士，金融基础弱

- **第一年**：把 CAPM、多因子、衍生品定价快速过完（一周 / 个）
- 优势：编程能力扎实，目标是头部私募的"量化开发"或"AI 研究员"
- **避免**：装作金融科班——直接说"我是 CS 背景但对金融感兴趣"反而加分

### 10.3 我是数学 / 物理 / 统计博士

- 你是私募最爱的人，竞争最激烈也最有优势
- 直投头部私募研究员，**不要走中型**
- 如果博士论文是 ML 方向，特别加分

### 10.4 我是工作几年想转量化的工程师

- 优势：工程能力 + 实战经验
- 路径：找需要工程能力的岗位（量化开发、数据平台、交易系统）
- 不要试图直接转研究员——和应届硕士竞争输面太大

### 10.5 我想去国外做量化

- 美国：纽约（Citadel、Two Sigma、DE Shaw）+ 芝加哥（Citadel Securities、Jump、HRT）
- 欧洲：伦敦（Two Sigma、Citadel、Man Group、AHL）+ 阿姆斯特丹（Optiver、IMC）
- 香港 / 新加坡：DRW、Optiver、Jane Street、Citadel
- 准备：刷 LC、Glassdoor 看面经、找校友内推
- 时间线：比国内提前 6 个月开始准备

## 11. 入职后怎么继续成长

### 11.1 头 6 个月

- 把公司的研究平台、回测框架、因子库摸熟
- 复刻一个已有策略，理解它的所有细节
- 至少读 20 篇核心论文（导师会推）
- **不要急着发表自己的"奇思妙想"**，95% 都是别人试过的

### 11.2 第 1 年

- 独立负责一个小策略 / 因子组
- 写至少 5 篇内部研究报告
- 开始关注组合层面、风险归因
- 主动跨组交流（数据、IT、交易）

### 11.3 第 2-3 年

- 成为某个细分方向的"小专家"
- 带 1-2 个实习生
- 跟着 senior 学组合管理、风控

### 11.4 永远在更新的知识

- arxiv.org/q-fin 每周扫一遍
- SSRN 关注几个核心作者
- 跟踪头部基金公开的研究（如 AQR、Bridgewater 的 daily observation）
- Twitter / X 上的 quant fintwit（注意鉴别垃圾）
