## 1. L2 (Layer 2)：以太坊扩容的现状

### 1.1 为什么需要 L2

以太坊主网瓶颈：
- 每秒最多 15-30 笔交易
- 简单转账 Gas 费 $5-20
- 复杂 DeFi 交互 $20-100+

L2 思想：**把计算放到链下，把结果摘要写回主网**。

主网角色：
- 数据可用性（DA）
- 最终确定性
- 安全（继承以太坊安全性）

L2 角色：
- 大量并行处理交易
- 几乎零 Gas（< $0.01）
- 高 TPS (1000+)

### 1.2 两大 L2 流派

**Optimistic Rollup（乐观汇总）**：
- 假设交易都是对的（乐观）
- 7 天挑战期：任何人可以证明交易非法
- 代表：Arbitrum、Optimism、Base、Blast

**ZK Rollup（零知识汇总）**：
- 每笔交易附带零知识证明
- 数学保证正确性，立即终局
- 代表：zkSync Era、StarkNet、Polygon zkEVM、Scroll、Linea

### 1.3 主要 L2 对比

| L2 | 类型 | TVL（峰值）| 特点 |
|---|---|---|---|
| **Arbitrum** | OP | $20B+ | TVL 最大，DeFi 生态最全 |
| **Base** (Coinbase) | OP | $15B+ | Coinbase 流量入口，新用户多 |
| **Optimism** | OP | $7B+ | Superchain 生态主导 |
| **Blast** | OP | $2B+ | 原生收益（ETH/USDB 自动生息）|
| **zkSync Era** | ZK | $1B+ | 早期 ZK，开发者体验好 |
| **StarkNet** | ZK | $1B+ | 自有 VM (Cairo)，长期愿景 |
| **Linea** (Consensys) | ZK | $1B+ | MetaMask 内置入口 |
| **Scroll** | ZK | $500M+ | 完全 EVM 兼容的 zkEVM |

### 1.4 选择 L2 的考虑

**做 DeFi 用户**：Arbitrum / Base（生态全、流动性好）
**做开发者**：先 Base + Optimism（同一 Superchain，部署一次复用代码）
**做长期投资**：ZK 系（技术更优，但生态早期）
**追新协议机会**：Blast / Mantle / zkSync 这种新链

### 1.5 ZK Rollup 的关键术语

| 术语 | 含义 |
|---|---|
| zkEVM | 与 EVM 兼容的 ZK 系统 |
| zk-SNARK | 简洁非交互式知识论证 |
| zk-STARK | 透明、抗量子 |
| Validity Proof | ZK 用的"对错证明" |
| Fraud Proof | OP 用的"欺诈证明" |
| Recursive Proof | 多个 proof 折叠成一个（zkSync 用）|

## 2. Account Abstraction (ERC-4337)：钱包的未来

### 2.1 传统 EOA（外部账户）的问题

| 问题 | 表现 |
|---|---|
| 助记词体验差 | 12 个英文单词，丢了找不回 |
| 必须有 ETH 付 Gas | 钱包 0 ETH 就废了 |
| 单签 = 单点风险 | 私钥被盗，全部资产丢 |
| 签名是黑盒 | 用户看不懂在签什么 |

### 2.2 AA 的本质

**把"账户"从一对密钥升级为一个智能合约**。

合约账户能：
- 自定义签名验证（多签、社交恢复、生物识别）
- 自定义 Gas 支付（用 USDC 付、其他人代付）
- 批量交易（一次签名多个操作）
- 时间锁、限额、权限分级

### 2.3 ERC-4337 架构

```
┌──── 用户 ────┐
│  发起一个 UserOp│
└──────┬───────┘
       │
       ▼
┌──── Bundler ────┐
│ 收集 UserOp     │
│ 打包成 tx        │
└──────┬──────────┘
       │
       ▼
┌──── EntryPoint 合约 ────┐
│ 验证签名（调用合约）    │
│ 执行 UserOp 内容        │
└─────────────────────────┘
       │
       ▼
   合约账户
```

**没有改以太坊协议，纯智能合约实现**。

### 2.4 几个真实使用案例

**Argent 钱包**：社交恢复（朋友 3/5 多签恢复）
**Safe (Gnosis Safe)**：多签钱包，DAO 标配
**Coinbase Smart Wallet**：邮箱即钱包，passkey 签名
**Soul Wallet**：智能合约钱包，Gas Sponsored

### 2.5 对量化交易者的意义

**实操好处**：
- 一个 ERC-4337 钱包可以**预设规则**：每天 1000 USDC 自动 swap、达到价位自动平仓
- **session keys**：临时授权某个 dApp 一段时间内可签名小额交易（DeFi 频繁交互不用每次确认）
- **gasless tx**：链上做市 / 套利时，可以用 USDC 付 Gas

**对 MEV 的影响**：
- AA 钱包可以原生集成 Flashbots Protect
- 多签 + 限额可以减少私钥泄露损失

## 3. Intent-Based Architecture：DEX 的范式转变

### 3.1 传统 DEX 的问题

传统 swap：
```
用户：我要在 Uniswap 用 X USDC 换 Y ETH
→ 用户自己选路径、自己付 Gas、自己承担滑点
→ Bot 看到这笔单去 mempool 抢跑/三明治
```

### 3.2 Intent-Based 的思想

**用户只表达意图**：

```
用户：我想用 1000 USDC 换 ETH，可以接受最低 0.49 ETH
→ 用户签名一个"意图"（Intent），不是具体的交易
→ Solver / Filler 网络竞争满足这个意图
→ 谁能给用户最好的报价，谁就赢得这单
```

### 3.3 代表项目

**UniswapX**（2023）：
- Uniswap 团队自己出的 intent 系统
- 链下订单簿 + 链上结算
- Solver 帮你找路径（可能用 Uniswap V3 + 1inch + 私人池子组合）

**CoW Swap**（前身 Gnosis Protocol）：
- "Coincidence of Wants"：如果两个用户互相想要对方的代币，直接 P2P 撮合，**零滑点 + 零 MEV**
- 批量拍卖：所有人同一时刻成交同一价格

**1inch Fusion / 0x**：
- 类似机制，整合多个 DEX

### 3.4 Intent 对生态的影响

**对普通用户**：
- 更好的成交价（多个 solver 竞争）
- 没有 MEV 损失（链下匹配）
- Gasless（Solver 替你付 Gas）

**对 MEV 行业**：
- Sandwich 没机会了（订单不上 mempool）
- 但产生了**新的 MEV**："Solver 之间的竞争"，类似 PBS

**对链上量化**：
- 写做市 / 套利 bot **必须改造成 Solver**
- 进入门槛提高（需要快速回应 + 集成多家 intent 协议）

### 3.5 一个 UniswapX 订单大概长什么样

```json
{
  "swapper": "0xUser...",
  "input": { "token": "USDC", "amount": "1000000000" },
  "output": {
    "token": "WETH",
    "startAmount": "498000000000000000",   // 0.498 ETH
    "endAmount":   "490000000000000000",   // 0.490 ETH（线性下降）
    "decayStart": 1730000000,
    "decayEnd":   1730000300                // 5 分钟拍卖
  },
  "exclusiveFiller": "0xPreferredSolver...",
  "signature": "0x..."
}
```

→ Solver 看到这个订单，从 1 分 30 秒后开始可以成交（价格逐渐下降到对自己有利）

## 4. EigenLayer + Restaking：再质押革命

### 4.1 问题：以太坊的"安全过剩"

ETH 质押有 ~$80B，但很多新协议没办法获得这种安全性。

### 4.2 EigenLayer 的天才想法

让以太坊验证者**复用质押的 ETH** 来给其他协议提供安全保障。

```
原本: 32 ETH 质押 → 只保护以太坊
现在: 32 ETH 质押 → 同时保护以太坊 + Service A + Service B + ...
        ↑ 多赚一份收益，但承担额外 slashing 风险
```

### 4.3 AVS（Actively Validated Services）

接受 restaked ETH 保护的"服务"。代表：

| AVS | 做什么 |
|---|---|
| **EigenDA** | 数据可用性（以太坊原生 DA 的便宜替代）|
| **Lagrange** | ZK 协处理器（链下计算 + 链上证明）|
| **Espresso** | 共享排序器（多 L2 共享）|
| **AltLayer** | rollup-as-a-service |
| **Witness Chain** | 物理基础设施验证 |

### 4.4 LRT（Liquid Restaking Token）

类似 stETH，但是 restaked 版本：

| LRT | 发行方 | TVL |
|---|---|---|
| eETH | EtherFi | $8B+ |
| ezETH | Renzo | $3B+ |
| rsETH | Kelp | $2B+ |
| pufETH | Puffer | $2B+ |

LRT 持有者：
- 赚 ETH 基础质押收益 (3-4%)
- 赚 EigenLayer points / AVS 奖励 (2-10%+)
- 流动性（可以 swap，可以做 LP）

### 4.5 风险

- **Slashing 级联**：一个 AVS 出问题导致大批验证者被 slash
- **去中心化退化**：所有 AVS 都依赖同一批验证者
- **复杂收益不可预测**：很多 LRT 收益"看起来诱人"实际是 token 通胀

### 4.6 对量化的机会

- **LRT 套利**：不同 LRT 之间价差，DEX 上经常有 30-100 bp 偏离
- **LRT 折溢价**：eETH/ETH peg 偏离的均值回归
- **EigenLayer points 农场化**：早期积分套利

## 5. 永续合约 DEX 大战

### 5.1 永续合约（Perp）是什么

类比传统期货，**没有到期日**（perpetual）的合约。

机制：通过"资金费率（Funding Rate）"让多空保持平衡，价格锚定现货。

- 多头资金费 > 0 → 多头给空头付钱（牛市常态）
- 多头资金费 < 0 → 空头给多头付钱（熊市常态）

### 5.2 五大永续 DEX 对比

| 协议 | 模式 | 日均交易量 | 特点 |
|---|---|---|---|
| **Hyperliquid** | 自建 L1 + 订单簿 | $5B+ | 速度极快，UI 流畅，最热门 |
| **dYdX V4** | Cosmos appchain | $1B+ | 完全去中心化订单簿 |
| **GMX V2** | AMM + 流动性池 | $500M+ | 适合大户，LP 角度好 |
| **Drift** | Solana 上 | $500M+ | Solana 最大永续 |
| **Vertex / Apex** | 混合模式 | $200M+ | 多模型混合 |

### 5.3 Hyperliquid 的颠覆性

**2024-2025 现象级项目**：

- 自建 L1（不是 L2），出块速度 0.1s
- 100% 链上订单簿（不是 AMM）
- 0 Gas（用户不付 Gas）
- 性能接近 Binance，开放性远超 CEX
- 2024 年 11 月空投 $HYPE，市值峰值 $30B

→ **"链上 Binance 时代"的开始**

### 5.4 永续 DEX 的链上量化机会

| 策略 | 描述 |
|---|---|
| **资金费套利** | 不同 perp 之间资金费率差异 |
| **CEX-DEX 价差** | Binance 现货 vs DEX 永续 |
| **做市** | 在永续订单簿上提供流动性 |
| **基差交易** | 现货多 + 永续空，赚资金费 |
| **跨协议套利** | dYdX vs Hyperliquid 价差 |

### 5.5 真实从业者收入

链上永续做市 / 套利 bot 团队：
- 小团队（3-5 人）年净利润 $1M - $10M 常见
- 顶级团队（10-30 人）$50M - $500M
- **进入门槛**：需要 CEX 量化 + Solidity + 工程 全才

## 6. 模块化区块链：未来的方向

### 6.1 一条区块链的四个功能层

```
┌─ Execution（执行）  EVM/SVM/MoveVM 跑智能合约
├─ Settlement（结算） 验证 proof 和欺诈证明
├─ Consensus（共识）  对块顺序达成共识  
└─ Data Availability（数据可用性） 数据公开可查
```

### 6.2 单体 vs 模块化

**单体链**：四个层一起，由同一组节点完成（Bitcoin、早期 Ethereum、Solana）

**模块化链**：拆开，每层可独立优化

```
你的应用 (Execution)
   ↓ 把数据存到
[Celestia / EigenDA]  ← 专门做 DA 的网络，便宜 100x
   ↓ 验证 proof 用
[Ethereum]            ← 安全性最高
```

### 6.3 关键项目

**Celestia**：
- 第一个专做 DA 的链
- L2 / L3 把数据扔过来，便宜
- 主网 2023 年 10 月上线，目前 TVL 不大但叙事强

**EigenDA**：
- 基于 EigenLayer 的 DA
- 用以太坊 restaked ETH 做安全保证
- 比 Celestia 更"以太坊原生"

**Avail**（Polygon）：
- Polygon 推出的 DA 链
- 已上线

**EigenLayer + Eigen DA + AltLayer**：
- 整套"模块化 rollup-as-a-service"栈
- 任何人花几小时部署自己的 rollup

### 6.4 对量化的影响

- **链多了**：未来跨链套利机会更多
- **执行环境多样化**：除了 EVM 还有 Move (Aptos/Sui)、SVM (Solana)、MoveVM
- **统一标准困难**：每个新执行环境都要重写代码

**做链上量化必须保持对新生态的探索**——错过 Solana 上一波 meme 的人都明白这点。

## 7. AI Agent + Crypto：2024-2025 的新叙事

### 7.1 什么是 AI Agent

**Agent = 能自主完成多步任务的 AI**（如 AutoGPT、Devin、Claude with Computer Use）

### 7.2 为什么 AI Agent 需要区块链

- **支付**：Agent 需要无人监管地支付（API 费、计算费）→ crypto 钱包
- **身份**：Agent 需要去中心化身份 → ENS / DID
- **协调**：多 Agent 协作需要可信中间层 → 智能合约
- **激励**：Agent 创造价值需要分发奖励 → tokens

### 7.3 代表项目

| 项目 | 做什么 |
|---|---|
| **Bittensor** | 去中心化 AI 训练网络，用 token 激励 |
| **Virtuals Protocol** | AI Agent 创建 + 代币化平台 |
| **Truth Terminal / GOAT** | 2024 年现象：AI Agent 在 X 上推出自己的 meme |
| **ai16z** | "AI 风投"概念项目 |
| **Fetch.ai** | 老牌 AI Agent + Crypto |
| **Olas** | Agent 协调网络 |

### 7.4 现实评估

**目前 99% 是炒作**：
- 大部分项目"AI"成分极弱，主要是 meme + 代币
- 真正的 AI Agent 还需要 2-3 年成熟

**但有价值的方向**：
- AI Agent 自动管理 DeFi 仓位
- AI Agent 做交易（早期实验）
- AI Agent 内容生成 + on-chain 经济激励

### 7.5 如果你想做 AI + Crypto

- 学好 ML + Solidity 双技能（稀缺）
- 关注 ZKML（下一节讲）：让 AI 计算的"诚实性"可验证
- 不要 FOMO 任何"AI Agent meme"——99% 归零

### 7.6 链上 LLM API 调用

一个有趣的新基础设施：

```solidity
// 智能合约里调用 LLM
function analyze() public {
    bytes32 requestId = oracleAI.requestInference(
        "claude-3.5-sonnet",
        "Analyze this news: ..."
    );
}
```

→ **DAO 治理**、**保险定价**、**链上仲裁**都可能被 LLM 改写。
